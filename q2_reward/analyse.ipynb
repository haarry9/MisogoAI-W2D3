## 📜 STEP 1: Define Prompts

```python
# 🟦 Cell 2: Define 5 prompts
prompts = [
    "Tell me a joke about programmers.",
    "Summarize the story of Cinderella.",
    "What is the meaning of life?",
    "Why is the sky blue?",
    "Write a short poem about the ocean."
]
```

## 🧠 STEP 2: Generate 4 Candidate Answers per Prompt

```python
# 🟩 Cell 3: Load GPT-2 model and tokenizer
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()

def generate_answers(prompt, num_answers=4):
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids
    outputs = model.generate(
        input_ids,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        num_return_sequences=num_answers,
        max_length=100,
        pad_token_id=tokenizer.eos_token_id
    )
    return [tokenizer.decode(out, skip_special_tokens=True) for out in outputs]
```

## 🎯 STEP 5: Load Dataset and Preprocess

```python
# ✅ Upload and clean the CSV
import pandas as pd
from datasets import Dataset

# Upload file first via files.upload() or drag-n-drop
df = pd.read_csv("answers.csv")

# Drop rows with missing values
df = df.dropna(subset=["prompt", "answer"])

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(df)
```

```python
# ✅ Fix tokenizer padding
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token

# Tokenize examples
def tokenize(example):
    return tokenizer(
        example["prompt"] + " " + example["answer"],
        padding="max_length",
        truncation=True,
        max_length=256,
    )

# Map and format
dataset = dataset.map(tokenize)
dataset = dataset.rename_column("rank", "labels")
dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

## 🏋️ STEP 6: Train Reward Model

```python
# 🟥 Cell 7: Load reward model (GPT-2 with classification head)
from transformers import AutoModelForSequenceClassification, TrainingArguments
from trl import RewardTrainer

reward_model = AutoModelForSequenceClassification.from_pretrained(
    "gpt2", num_labels=1
)
reward_model.resize_token_embeddings(len(tokenizer))
```

```python
# 🟧 Cell 8: Training config
training_args = TrainingArguments(
    output_dir="reward_model",
    per_device_train_batch_size=2,
    num_train_epochs=1,
    max_steps=100,
    logging_steps=10,
    save_steps=50,
    save_total_limit=1,
    report_to="none"
)

# Explicitly set disable_dropout to False
setattr(training_args, 'disable_dropout', False)

trainer = RewardTrainer(
    model=reward_model,
    args=training_args,
    train_dataset=dataset,
    processing_class=tokenizer,
)

trainer.train()
```

## STEP 7: Evaluate with New Answers and Plot

```python
# 🟦 Cell 9: Predict reward scores for new answers
import matplotlib.pyplot as plt

test_prompt = "Why do cats purr?"
test_answers = generate_answers(test_prompt)

inputs = tokenizer([test_prompt + " " + ans for ans in test_answers],
                   return_tensors="pt", padding=True, truncation=True)

with torch.no_grad():
    rewards = reward_model(**inputs).logits.squeeze().tolist()

# Plot scores
plt.bar(range(1, 5), rewards)
plt.xlabel("Answer #")
plt.ylabel("Predicted Reward")
plt.title("Reward Scores for Test Answers")
plt.show()

for i, (a, s) in enumerate(zip(test_answers, rewards), 1):
    print(f"Answer {i} (Score: {s:.2f}):\n{a.strip()}\n")
```
